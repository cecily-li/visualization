```python
#Import packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

#Assign random seed for reproducibility
np.random.seed(613)

#Arrange 50 data points on y-axis between 0 and 100, x-axis between 0 and 50
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
#Create a line graph for both y variables
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
#Plot y1 variable "Person 1", y2 variable "Person 2" and create a legend to be displayed at the lower right location
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right'
,
frameon = True, #add frame around the legend
fontsize = 12, #change font size
ncol = 2, #specify number of columns to present variable names in
shadow = True ) #add drop shadow around legend
fig.show()
```

```python
#Position text annotations using transforms
#Create x and y axes with ranges 0 to 10
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

#1 over, 5 up
ax.text(1,5,"DataL(1,5)", transform = ax.transData)
#5 over, 1 up on axes
ax.text(0.5,0.1,"Axes:(0.5,0.1)", transform = ax.transAxes)
#2 over, 2 up on figure dimension
ax.text(0.2,0.2,"Figure:(0.2,0.2)", transform = fig.transFigure)
```

```python
#Annotate with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#Point arrow at coordinates (10,95), put text at (20,94)
ax.annotate('This is important!'
, xy=(10, 95), xytext=(20, 94),
arrowprops=dict(facecolor='black' #face colour of arrow is black
, shrink=0.05)) #shrink value determines how far that arrow shrinks to/from where it's pointing
fig.show()
```

```python
#Annotate with arrows
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
#Point arrow at coordinates (10,95), put text at (20,94)

ax.annotate("This is important!"
,
xy=(10, 95), xytext=(20, 94),
arrowprops = dict(arrowstyle = "wedge" #change arrow style to wedge from pointer arrow
,
color = "hotpink")) #change arrow colour to hot pink
fig.show()
```

```python
#Remove ticks and labels from scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right') #add in legend at lower right
ax.yaxis.set_major_locator(plt.NullLocator()) #set major locator for y axis using null locator (remove ticks and labels)
ax.xaxis.set_major_formatter(plt.NullFormatter()) #set major formatter for x axis using null formatter (removes labels)
```

```python
#Only show 3 ticks on the x axis using MaxNLocator
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
#Use Multiple locator to set tickmarks to appear at a particular interval (i.e., every 5 unit interval on x axis)
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```

```python
#Rotate the way axis labels are presented
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc="lower right")
#Rotate x tickmarks 45 degrees to the right, horizontal alignment to the right
plt.xticks(rotation=45, ha="right")
```

```python
#Generate random data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#Define 2 axes to have 2 plots side by side
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(7, 3))

ax1.scatter(x1,y1,
            marker='o',
            color = 'red') #scatter plot
ax2.bar(x2,y2,
        color = 'green') #bar plot
fig.show()
```

```python
#Listing numerous axes as labelled (ax1, ax2, etc.)
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
#Listing numerous axes as labelled (ax1, ax2, etc.)
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes['ax1'].set_xlabel("Another Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
```

```python
#Listing numerous axes as labelled (ax1, ax2, etc.) with constraints
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize=(7, 4),
                                    layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes['ax1'].set_xlabel("Another Label", fontsize=18)
someaxes["ax2"].set_xlabel("Another Label", fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious", fontsize=18)
```

```python
# first make our sample data
x = np.array(["Luffy", "Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])
# define figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))
#call both bar and plot elements to the same axes (ax)
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2,
color = "red")
```

```python
#Calculate sd of y2
y2_sd = np.std(y2)

#plot data as normal 
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

#add vertical error bars
ax.errorbar(x,y2,
            yerr = y2_sd,
            fmt="none",
            ecolor="indigo", #change color of error bar
            elinewidth=4, #change thickness of error bar
            capsize=6, #add cap on error bar
            capthick=4, #change thickness of cap
            errorevery=2) #show every other error bar
```

```python
#Import packages
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

```

```python
df = pd.read_csv('Apartment Building Evaluations 2023 - current.csv')
df.info()
```

```python
df.select_dtypes("number")
```

```python
y = df['CURRENT BUILDING EVAL SCORE']
x1 = df['YEAR BUILT']
x2 = df ['CONFIRMED STOREYS']
#fig, (ax1,ax2) = plt.subplots(1,2,figsize=(25, 18))

#ax1.scatter(x1, y,
#color = "indigo")

#Define 2 axes to have 2 plots side by side
fig,(ax1, ax2) = plt.subplots(ncols=2,
nrows=1,
figsize=(30, 16))

ax1.scatter(x1,y,
            marker='o',
            color = 'blue') #scatter plot of building evaluation score against year the apartment was built (Figure 1)
ax2.scatter(x2,y,
        color = 'black') #scatter plot of building evaluation score against the number of storeys of the building (Figure 2)

#Update axes labels and title for figure 1
ax1.set_xlabel('Year Built')
ax1.set_ylabel('Building Evaluation Score')
ax1.set_title('Figure 1: Building Evaluation Score by Year Built')

#Update axes labels and title for figure 2
ax2.set_xlabel('Number of Storeys')
ax2.set_ylabel('Building Evaluation Score')
ax2.set_title('Figure 2: Building Evaluation Score by Number of Storeys')
```
